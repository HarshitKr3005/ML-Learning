```python
import pandas as pd
data = pd.read_csv(r'C:\Users\harsh\Desktop\Machine Learning\Datasets\tennis.csv')
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
outlook=LabelEncoder()
temp=LabelEncoder()
humidity=LabelEncoder()
windy=LabelEncoder()
play=LabelEncoder()
```

```python
data['outlook']= outlook.fit_transform(data['outlook'])
data['temp']= outlook.fit_transform(data['temp'])
data['humidity']= outlook.fit_transform(data['humidity'])
data['windy']= outlook.fit_transform(data['windy'])
data['play']= outlook.fit_transform(data['play'])
```

```python
data
```

```python
features = ['outlook','temp','humidity','windy']
x = data[features]
y = data.play
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy')
classifier.fit(x_train,y_train)
```

```python
classifier.predict(x_test)
```

```python
x_test
```

```python
classifier.score(x_test,y_test)
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```
